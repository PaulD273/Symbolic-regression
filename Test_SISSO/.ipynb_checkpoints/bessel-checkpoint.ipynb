## Max_rung = 3

```python
x0 = symbols('x')
c0 = 9.986912122335090e-01
a0 = -2.368620197301263e-01
# Define the expression
expr = c0 + a0 * ((((x0**2) - x0) * ((x0**2) + x0)) / (x0**2))
expand(expr)
```

## Max_rung = 4

```python
x0 = symbols('x')
c0 = 7.618291925033870e-01
a0 = -2.368620197301215e-01
# Define the expression
expr = c0 + a0 * ((((x0**2) - x0) * ((x0**2) + x0)) / (x0**2))
expand(expr)
```

## Max_rung = 5

```python
feature_0 = symbols('x')
c0 = -6.708169940169566e-02
a0 = 1.001205511353395e+00
# Define the expression
expr = c0 + a0 * (((((feature_0**2)**2)**2) / (((feature_0**2)**2) + (feature_0**2))) - ((((feature_0**2) - feature_0) - feature_0) * (((feature_0**2) + feature_0) + feature_0)))
expand(expr)
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 500)
sissopp_target3 = 9.986912122335090e-01 -2.368620197301263e-01*x**2
sissopp_target4 = 7.618291925033870e-01 -2.368620197301215e-01 * ((((x**2) - x) * ((x**2) + x)) / (x**2))
sissopp_target5 = -6.708169940169566e-02 + 1.001205511353395e+00 * (((((x**2)**2)**2) / (((x**2)**2) + (x**2))) - ((((x**2) - x) - x) * (((x**2) + x) + x)))
pysr_target = -0.000419888994504707 * x**6 + 0.015615513 * x**4 - 0.24999791 * x**2 + 0.99999994
```

```python
plt.scatter(x, pysr_target, s = 20, label = 'PySR result with niter = 300')
plt.scatter(x, sissopp_target3, s = 20, label = 'SISSO++ result with max_rung=3')
plt.scatter(x, sissopp_target4, s = 20, label = 'SISSO++ result with max_rung=4')
plt.scatter(x, sissopp_target5, s = 20, label = 'SISSO++ result with max_rung=5')
plt.plot(x, jn(0,x), color='r', label = 'Ideal function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.ylim([-10,2])
#plt.xlim([-2,2])
plt.legend()
plt.show()
```
