```python
arr = np.array(data[:800])
arr[:3], arr.shape
```

```python
# Dataset
X = arr[:, :9]
m1 = X[:,0]
m2 = X[:,1]
G = X[:,2]
x1 = X[:,3]
x2 = X[:,4]
y1 = X[:,5]
y2 = X[:,6]
z1 = X[:,7]
z2 = X[:,8]
#theta, sigma
y = arr[:, 9]
```

```python
r = np.sqrt((x1-x2)**2 + (y1-y2)**2  + (z1-z2)**2)
red_X = np.vstack((m1, m2, G, r)).T
```
