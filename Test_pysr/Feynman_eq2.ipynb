```python
arr = np.array(data[:1000])
arr[:3], arr.shape
```

```python
# Dataset
X = arr[:, :4]
h = X[:,0]
omega = X[:,1]
kb = X[:,2]
T = X[:,3]
y = arr[:, 4]
```

```python
# Learn equations
model = PySRRegressor(
    niterations=600,
    binary_operators=["*", "/", "+", "-", "^"],
    unary_operators=["exp"],
    #extra_sympy_mappings={"ekbT": lambda x, y: np.exp(1/(x*y))},
    **default_pysr_params,
)

model.fit(X, y, variable_names=["h", "omega", "kb", "T"])
```
