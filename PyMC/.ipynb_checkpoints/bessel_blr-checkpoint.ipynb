```python
import pymc as pm
import arviz as az
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sympy as sp
from io import StringIO
from scipy.special import jn
```

```python
# Read the file
with open('./Rankings/bessel.txt', 'r') as f:
    lines = f.readlines()

# Remove the header (starts with '#')
lines = [line for line in lines if not line.strip().startswith('#')]

# Join the cleaned lines into a single string
data = ''.join(lines)

# Use pandas to read from the cleaned string and converts it into a DataFrame
df = pd.read_csv(StringIO(data), sep=r'\s{2,}', engine='python', header=None)
df.columns = ['FEAT_ID', 'Score', 'Feature Expression']

print(df)
```

```python
x = sp.symbols('feature_0')
symbol_map = {'x': x}
formulas = df['Feature Expression']

# Converts the equtions (stored into the 'Feature Expression' column of df) from strings into sympy objects
sympy_exprs = [sp.sympify(f, locals=symbol_map) for f in formulas]
# Builds Python functions starting from the sympy expressions
functions = [sp.lambdify((x), expr, modules='numpy') for expr in sympy_exprs]
```

```python
%config InlineBackend.figure_format = 'retina'
# Initialize random number generator
RANDOM_SEED = 8927
rng = np.random.default_rng(RANDOM_SEED)
az.style.use("arviz-darkgrid")

n_samples = 300
# Defines a training set
x_train = np.linspace(-2, 2, n_samples)
# Builds an object Phi by stacking the Python functions defined previously, evaluated at x_train
Phi = np.column_stack([f(x_train) for f in functions])

n_features = Phi.shape[1]

# Define the target function and adds a noise for each point separately using size
y = jn(0,x_train) + rng.normal(size=x_train.shape)
```

```python
basic_model = pm.Model()

with basic_model:
    # Priors for unknown model parameters
    # Normal weights
    w = pm.Normal("w", mu=0, sigma=10, shape=n_features)
    # Sigma is positively define and normally distributed --> HalfNormal
    sigma = pm.HalfNormal("sigma", sigma=1)

    # Expected value of outcome, Linear model
    mu = pm.math.dot(Phi, w)

    # Likelihood (sampling distribution) of observations --> normal distribution with the deterministic equation as mean and sigma as variance
    y_obs = pm.Normal("y_obs", mu=mu, sigma=sigma, observed=y)
```

```python
with basic_model:
    # To change the sampling method, NUTS should be the default
    #step = pm.Metropolis()
    # draw 1000 posterior samples from the unnormalized posterior distribution computed as likelihood x prior
    idata = pm.sample()
```

```python
#az.plot_trace(idata, combined=True);
```

```python
# Predictive data
# defines a test set, including points which lie outside the training set, and evaluate Phi at these points
x_test = np.linspace(-2.1, 2.1, 300)
Phi_test = np.column_stack([f(x_test) for f in functions])

# Get posterior samples of weights and sigma 
posterior_w = idata.posterior["w"].stack(draws=("chain", "draw")).values  # (n_features, n_draws)
posterior_sigma = idata.posterior["sigma"].stack(draws=("chain", "draw")).values  # (n_draws,)

# Compute predictions
y_preds = posterior_w.T @ Phi_test.T  # shape: (n_draws, n_test)
# Numerical expectation of y given x and the dataset weighted by the posterior distribution --> Posterior predictive distribution
y_mean = y_preds.mean(axis=0)
y_std = y_preds.std(axis=0)
```

$$
    E(y | x, D) = \int \Phi(x) \ w \ p(w | D) \ dw \sim \frac{1}{N} \sum_{i=1}^N \Phi(x) \cdot w
$$
Approximated as a running average of the predictions

```python
plt.figure(figsize=(10, 6))
plt.plot(x_train, y, "o", label="Test points", alpha=0.5)
plt.plot(x_test, jn(0,x_test), label="True Function", color="green", linestyle="--")
plt.plot(x_test, y_mean, label="Predicted Mean", color="blue")
plt.fill_between(x_test, y_mean - 2*y_std, y_mean + 2*y_std, alpha=0.3, color="blue", label=r"±2$\sigma$ Uncertainty")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Bayesian Linear Regression with SISSO++ Features")
plt.grid(True)
plt.show()
```
